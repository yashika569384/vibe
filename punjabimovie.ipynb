```python
import pandas as pd
df7=pd.read_csv('punjmov.csv',encoding='ISO-8859-1')
df7
```

```python
df7['combined_features']=df7['genres']+df7['original_language']
df7['combined_features']
```

```python
ds7  = df7.drop(columns=['genres','original_language','poster_path'])
ds7
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv7=CountVectorizer(max_features=10000, stop_words='english')
c7=cv7.fit_transform(df7['combined_features'].values.astype('U')).toarray()
```

```python
c7.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c7)
similarity
```

```python
idx7=ds7[ds7['Title']=="Sajjan Singh Rangroot"].index[0]
idx7
```

```python
import random
distance7 = sorted(list(enumerate(similarity[idx7])), reverse=True, key=lambda vector:vector[1])

s7=[ds7.iloc[i[0]].Title for i in distance7[1:10]]
random.shuffle(s7)
for i in s7[:5]:
    print(i)
```

```python
def recommand(movie_name):
    index7=ds7[ds7['Title']==movie_name].index[0]
    distance7 = sorted(list(enumerate(similarity[index7])), reverse=True, key=lambda vector:vector[1])
    s7=[ds7.iloc[i[0]].Title for i in distance7[1:10]]
    random.shuffle(s7)
    for i in s7[0:5]:
        print(i)
```

```python
recommand("Sajjan Singh Rangroot")
```

```python
import pickle
pickle.dump(ds7, open('moviepunjabi.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.load(open('moviepunjabi.pkl', 'rb'))
```
