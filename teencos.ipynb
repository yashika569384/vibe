```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
dft=pd.read_csv('teens.csv',encoding='ISO-8859-1')
dft
```

```python
dft['combined_features']=dft['Genre']+dft['Rating']
dft['combined_features']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
c=cv.fit_transform(dft['combined_features'].values.astype('U')).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c)
```

```python
idx=dft[dft['Name']=="Star Wars The Clone Wars: Lightsaber Duels"].index[0]
idx
```

```python
import random
distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda vector:vector[1])
# 
unique_m = []  #list
for i in distance[1:10]:  # Skip the first one (itself) and take the next 9
    m_name = dft.iloc[i[0]].Name
    if m_name not in unique_m:  # Ensure uniqueness
        unique_m.append(m_name)

# Shuffle the unique movie names
random.shuffle(unique_m)
for i in unique_m[:5]:
    print(i)
```

```python
import pickle
pickle.dump(dft, open('teencos.pkl', 'wb'))
pickle.dump(similarity, open('teen.pkl', 'wb'))
pickle.load(open('teencos.pkl', 'rb'))
```
