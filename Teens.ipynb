```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
dft=pd.read_csv("teens.csv")
dft
```

```python
dt=dft
dt.head()
```

```python
dt['Rating'] = dt['Rating'].replace('E10+', 'Above 10 years')
dt
```

```python
dt['Rating'] = dt['Rating'].replace('T', 'Teens')
dt
```

```python
dt.head()
dt
```

```python
dt['combined_features']=dt['Genre']+dt['Rating']
dt['combined_features']
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(dt['combined_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
knn = NearestNeighbors(n_neighbors=10, metric='cosine')  # n_neighbors=10 means top-10 similar games
knn.fit(cosine_sim)
```

```python
def recommend_games(name):
    game_index = dt[dt['Name'].str.lower() == name.lower()].index[0]
    distance, indices = knn.kneighbors(cosine_sim[game_index].reshape(1, -1), n_neighbors=6)
    recommended_game_indices = indices.flatten()[1:]
    recommended_games = dt.iloc[recommended_game_indices]
    return recommended_games['Name']

g = recommend_games('Tekken 6')

# Print movie names without index
for m in g:
    print(m)
```

```python
import pickle
pickle.dump(dt, open('teen.pkl', 'wb'))
pickle.dump(cosine_sim, open('similarity.pkl', 'wb'))
pickle.load(open('teen.pkl', 'rb'))
```
