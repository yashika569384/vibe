```python
import pandas as pd
dfet=pd.read_csv('Everyone10+.csv', encoding='ISO-8859-1')
dfet
```

```python
dfet['combined_features']=dfet['Genre']+dfet['Rating']
dfet['combined_features']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
c=cv.fit_transform(dfet['combined_features'].values.astype('U')).toarray()
```

```python
c.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c)
similarity
```

```python
idx=dfet[dfet['Name']=="LEGO Batman: The Videogame"].index[0]
idx
```

```python
import random
distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda vector:vector[1])
# 
unique_et = []  #list
for i in distance[1:10]:  # Skip the first one (itself) and take the next 9
    et_name = dfet.iloc[i[0]].Name
    if et_name not in unique_et:  # Ensure uniqueness
        unique_et.append(et_name)

# Shuffle the unique movie names
random.shuffle(unique_et)
for i in unique_et[:5]:
    print(i)
```
