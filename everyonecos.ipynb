```python
import pandas as pd
dfe=pd.read_csv('everyone.csv',encoding='ISO-8859-1')
dfe
```

```python
dfe['combined_features']=dfe['Genre']+dfe['Rating']
dfe['combined_features']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
c=cv.fit_transform(dfe['combined_features'].values.astype('U')).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c)
```

```python
idx=dfe[dfe['Name']=="Bakugan: Battle Brawlers"].index[0]
idx
```

```python
import random
distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda vector:vector[1])
# 
unique_e = []  #list
for i in distance[1:10]:  # Skip the first one (itself) and take the next 9
    game_e = dfe.iloc[i[0]].Name
    if game_e not in unique_e:  # Ensure uniqueness
        unique_e.append(game_e)

# Shuffle the unique movie names
random.shuffle(unique_e)
for i in unique_e[:5]:
    print(i)
```

```python
import pickle
pickle.dump(dfe, open('e.pkl', 'wb'))
pickle.dump(similarity, open('every.pkl', 'wb'))
pickle.load(open('e.pkl', 'rb'))
```
