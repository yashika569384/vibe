```python
import pandas as pd
df1=pd.read_csv('englishmusic.csv')
df1
```

```python
df1.drop('duration',axis=1,inplace=True)
df1.drop('popularity',axis=1,inplace=True)
df1
```

```python
df1['combined_features']=df1['Genre']+df1['language']
df1['combined_features']
```

```python
ds1  = df1.drop(columns=['Genre','language','poster'])
ds1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
c=cv.fit_transform(df1['combined_features'].values.astype('U')).toarray()
```

```python
c.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c)
similarity
```

```python
idx1=ds1[ds1['song_name']=="closer"].index[0]
idx1
```

```python
import random
distance1 = sorted(list(enumerate(similarity[idx1])), reverse=True, key=lambda vector:vector[1])

s=[ds1.iloc[i[0]].song_name for i in distance1[1:10]]
random.shuffle(s)
for i in s[:5]:
    print(i)
```

```python
def recommand(song_name):
    index=ds1[ds1['song_name']==song_name].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    s1=[ds1.iloc[i[0]].song_name for i in distance[1:10]]
    random.shuffle(s1)
    for i in s1[0:5]:
        print(i)
```

```python
recommand("no one")
```
