```python
import pandas as pd
df6=pd.read_csv('engmov.csv')
df6
```

```python
df6['combined_features']=df6['genres']+df6['original_language']
df6['combined_features']
```

```python
ds6 = df6.drop(columns=['genres','original_language','poster_path'])
ds6
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv6=CountVectorizer(max_features=10000, stop_words='english')
c6=cv6.fit_transform(df6['combined_features'].values.astype('U')).toarray()
```

```python
c6.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c6)
similarity
```

```python
idx6=ds6[ds6['Title']=="The Bubble"].index[0]
idx6
```

```python
import random
distance6 = sorted(list(enumerate(similarity[idx6])), reverse=True, key=lambda vector:vector[1])

s6=[ds6.iloc[i[0]].Title for i in distance6[1:10]]
random.shuffle(s6)
for i in s6[:5]:
    print(i)
```

```python
def recommand(mov_name):
    index6=ds6[ds6['Title']==mov_name].index[0]
    distance6 = sorted(list(enumerate(similarity[index6])), reverse=True, key=lambda vector:vector[1])
    m1=[ds6.iloc[i[0]].Title for i in distance6[1:10]]
    random.shuffle(m1)
    for i in m1[0:5]:
        print(i)
```

```python
recommand("Father Stu")
```

```python
import pickle
pickle.dump(ds6, open('movieeng.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.load(open('movieeng.pkl', 'rb'))
```
