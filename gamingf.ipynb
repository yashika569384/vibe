```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
df=pd.read_csv("gamesff.csv")
df
```

```python
# df3=df.drop(['Platform','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales','Critic_Score','Critic_Count','User_Score','User_Count','Developer'],axis=1,inplace=True)
df3=df
df3.head()
```

```python
df3.dropna(inplace=True)
df3
```

```python
df3.info()
```

```python
df3['Rating'] = df3['Rating'].replace('E', 'Everyone')
df3
```

```python
df3['Rating'] = df3['Rating'].replace('E10+', 'Above 10 years')
df3
```

```python
df3['Rating'] = df3['Rating'].replace('T', 'Teens')
df3['Rating'].unique()
df3['Rating'] = df3['Rating'].replace('M', 'Mature')
df3
```

```python
df3['Rating'] = df3['Rating'].replace('K-A', 'From Kids To Adult')
df3['Rating'] = df3['Rating'].replace('AO', 'Adults Only')
df3['Rating'] = df3['Rating'].replace('EC', 'Below 10 Years')
df4=df3[df3["Rating"] != "RP"]
df4.head()
```

```python
df4
```

```python
df4['combined_features']=df4['Genre']+df4['Rating']
df4['combined_features']
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df4['combined_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
knn = NearestNeighbors(n_neighbors=10, metric='cosine')  # n_neighbors=10 means top-10 similar games
knn.fit(cosine_sim)
```

```python
def recommend_games(name):
    game_index = df4[df4['Name'].str.lower() == name.lower()].index[0]
    distance, indices = knn.kneighbors(cosine_sim[game_index].reshape(1, -1), n_neighbors=6)
    recommended_game_indices = indices.flatten()[1:]
    recommended_games = df4.iloc[recommended_game_indices]
    return recommended_games['Name']

g = recommend_games('Wii Play')

# Print movie names without index
for m in g:
    print(m)
```
