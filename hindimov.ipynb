```python
import pandas as pd
df5=pd.read_csv('hindimov.csv',encoding='ISO-8859-1')
df5
```

```python
df5['combined_features']=df5['genres']+df5['original_language']
df5['combined_features']
```

```python
ds5  = df5.drop(columns=['genres','original_language','poster_path'])
ds5
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv5=CountVectorizer(max_features=10000, stop_words='english')
c5=cv5.fit_transform(df5['combined_features'].values.astype('U')).toarray()
```

```python
c5.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c5)
similarity
```

```python
idx5=ds5[ds5['Title']=="Aashiqui 2"].index[0]
idx5
```

```python
import random
distance5 = sorted(list(enumerate(similarity[idx5])), reverse=True, key=lambda vector:vector[1])

s5=[ds5.iloc[i[0]].Title for i in distance5[1:10]]
random.shuffle(s5)
for i in s5[:5]:
    print(i)
```

```python
def recommand(movie_name):
    index5=ds5[ds5['Title']==movie_name].index[0]
    distance5 = sorted(list(enumerate(similarity[index5])), reverse=True, key=lambda vector:vector[1])
    s5=[ds5.iloc[i[0]].Title for i in distance5[1:10]]
    random.shuffle(s5)
    for i in s5[0:5]:
        print(i)
```

```python
recommand("Aashiqui 2")
```

```python
import pickle
pickle.dump(ds5, open('moviehindi.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.load(open('moviehindi.pkl', 'rb'))
```
