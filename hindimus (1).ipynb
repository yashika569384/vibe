```python
import pandas as pd
df2=pd.read_csv('hindimusic.csv',encoding='ISO-8859-1')
df2
```

```python
df2.drop('duration',axis=1,inplace=True)
df2.drop('popularity',axis=1,inplace=True)
df2
```

```python
df2['combined_features']=df2['Genre']+df2['language']
df2['combined_features']
```

```python
ds2  = df2.drop(columns=['Genre','language','poster'])
ds2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv2=CountVectorizer(max_features=10000, stop_words='english')
c2=cv2.fit_transform(df2['combined_features'].values.astype('U')).toarray()
```

```python
c2.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c2)
similarity                                
```

```python
idx2=ds2[ds2['song_name']=="qatra"].index[0]
idx2
```

```python
import random
distance2 = sorted(list(enumerate(similarity[idx2])), reverse=True, key=lambda vector:vector[1])

s2=[ds2.iloc[i[0]].song_name for i in distance2[1:10]]
random.shuffle(s2)
for i in s2[:5]:
    print(i)
```

```python
def recommand(song_name):
    index2=ds2[ds2['song_name']==song_name].index[0]
    distance2 = sorted(list(enumerate(similarity[index2])), reverse=True, key=lambda vector:vector[1])
    s1=[ds2.iloc[i[0]].song_name for i in distance2[1:10]]
    random.shuffle(s1)
    for i in s1[0:5]:
        print(i)
```

```python
recommand("qatra")
```

```python
import pickle
pickle.dump(ds2, open('musichindi.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.load(open('musichindi.pkl', 'rb'))
```
